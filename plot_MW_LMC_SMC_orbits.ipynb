```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
```

All data from figure 1 here: https://ui.adsabs.harvard.edu/abs/2020ApJ...893..121P/abstract

Note three different LMC masses: </br>
LMC1: 8e10 Msun </br>
LMC2: 18e10 Msun (** fiducial mass) </br> 
LMC3: 25e10 Msun </br>

Two different MW Masses: </br>
MW1: 1e12 Msun </br>
MW2: 1.5e12 Msun </br>

SMC1 mass is fixed at 5e9 Msun.

All orbits are 3-body envounters between the MW-LMC-SMC. Note that the MW is not fixed in space so all coordinates should be shifted to MW COM before using (depending on the application). 

Time units are all in Gyr and distances are all in kpc. Velocities are also included in these data files (columns 4,5,6 10,11,12,16,17,18) as km/s.


# load data for MW1 and LMC1

```python
data = np.loadtxt('MW1_LMC2_SMC1_orbit_mean_v2_600.txt')
ts = data[:,0] #Gyr
ts = [-t for t in ts]

# MW position components 
mxs = data[:,1]
mys = data[:,2]
mzs = data[:,3]

# LMC position components 
lxs = data[:,7]
lys = data[:,8]
lzs = data[:,9]

# SMC position components 
sxs = data[:,13]
sys = data[:,14]
szs = data[:,15]

#orbit of LMC-SMC
rs = [la.norm([x-x2,y-y2,z-z2]) for x,y,z,x2,y2,z2 in zip(sxs, sys, szs, lxs, lys, lzs)]

#orbit of SMC wrt MW
smcrs = [la.norm([x-x2,y-y2,z-z2]) for x,y,z,x2,y2,z2 in zip(sxs, sys, szs, mxs, mys, mzs)]

#orbit of LMC wrt MW
lmcrs = [la.norm([x-x2,y-y2,z-z2]) for x,y,z,x2,y2,z2 in zip(lxs, lys, lzs, mxs, mys, mzs)]
```

```python
plt.figure(figsize=(6,6))


#LMC1 wrt MW
plt.plot(ts, lmcrs, c='C4',label=r'LMC orbit around MW')

#SMC wrt MW
plt.plot(ts, smcrs, label='SMC orbit around MW')

#SMC wrt LMC1
plt.plot(ts, rs, color='C3', label='LMC-SMC orbit')

plt.axhline(y=261,ls=':', color='gray') #MW1 Rvir (MW2 Rvir = 299)
plt.ylim(0,400)
plt.text(-0.1,263, 'MW virial radius', color='gray')
plt.legend(loc='upper left', frameon=False)
plt.ylabel('relative distance [kpc]', labelpad=2.5)
plt.xlabel('lookback time [Gyr]')
```
